```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = "https://www.iplt20.com/auction/2022"
```

```python
r = requests.get(url)
print(r)
```

```python
r.content
```

```python
soup = bs(r.content,"html.parser")
print(soup.prettify())
```

```python
table = soup.find("table", class_ = "ih-td-tab auction-tbl")
print(table)
```

```python
# for getting the headers of tables
header = table.find_all("th")
print(header)
```

```python
# create the empty list to add different headers in it:
titles = []

#exract the headers:
for i in header:
    title = i.text
    titles.append(title)
print(titles)    
```

```python
# creating the dataframes:
df = pd.DataFrame(columns=titles)
print(df)
```

```python
# getting rows:
rows = table.find_all('tr')
for i in rows[1:]:      # use of [1:] so header not get printed
    data = i.find_all('td')
    row = [tr.text for tr in data] # all rows data taken with this
    l = len(df)  # help to form data in tabular form
    df.loc[l] = row
print(df)    
```

```python
df["TEAM"] = df["TEAM"].apply(lambda x:x.strip())
print(df)
```

```python
df.to_csv('Tata_acution_stats_2022.csv')
```
